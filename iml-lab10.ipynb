```python
import pandas as pd
df = pd.read_csv('/kaggle/input/nifty50-stock-market-data/TATAMOTORS.csv')
# Display the first 20 rows of the original dataset
print("Original Dataset (first 20 rows):")
print(df.head(20))
# Drop missing values
df = df.dropna()
# Remove duplicates
df = df.drop_duplicates()
# Display the first 20 rows of the cleaned dataset
print("\nCleaned Dataset (first 20 rows):")
print(df.head(20))
```
